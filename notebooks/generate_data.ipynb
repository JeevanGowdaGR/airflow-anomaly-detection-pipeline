```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Generate time-series data
n = 500
time = pd.date_range(start='2024-01-01', periods=n, freq='h')

values = np.random.normal(loc=50, scale=5, size=n)

# Inject anomalies
anomaly_indices = np.random.choice(n, size=20, replace=False)
values[anomaly_indices] += np.random.normal(30, 5, size=20)

df = pd.DataFrame({
    "timestamp": time,
    "value": values
})

# Save dataset
df.to_csv("../data/time_series.csv", index=False)

df.head()
```
